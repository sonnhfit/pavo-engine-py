```python
import pandas as pd
```

```python
a = pd.read_csv(r"/Users/admin/Downloads/abc.csv")
```

```python
a
```

```python
b = pd.read_csv(r"/Users/admin/Downloads/cde.csv")
```

```python
a['Cum. Profit'] = a['Cum. Profit'].str.replace(',', '.').str.rstrip('%').astype(float)
b['Cum. Profit'] = b['Cum. Profit'].str.replace(',', '.').str.rstrip('%').astype(float)
```

```python
import matplotlib.pyplot as plt

# Plot Cumulative Profit line chart for dataframe a
a['Cum. Profit'].plot(label='a')

# Plot Cumulative Profit line chart for dataframe b
b['Cum. Profit'].plot(label='b')

# Set the title and labels for the chart
plt.title('Cumulative Profit')
plt.xlabel('Index')
plt.ylabel('Cumulative Profit')

# Display the legend
plt.legend()

# Display the chart
plt.show()
```
