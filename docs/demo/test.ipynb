```python
#!/usr/bin/env python
import pika


```

```python
credentials = pika.PlainCredentials('admin', 'mypass')
connection = pika.BlockingConnection(pika.ConnectionParameters(host='0.0.0.0', credentials=credentials))
```

```python
channel = connection.channel()

channel.queue_declare(queue='hello')

channel.basic_publish(exchange='', routing_key='hello', body='{"name": "John", "age": 30, "city": "New York"}')
print(" [x] Sent 'Hello World!'")
```

```python
connection.close()
```
